```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset_barris.csv')
display(df)
```

```python
# Remove the unneeded columns for the model
df_cleaned = df.drop(columns=['synthetic_id', 'neighbourhood', 'address',
                              'districte', 'section', 'district_id', 'section_id',
                              'census_tract_IDESCAT'])
display(df_cleaned.head())
```

```python
# Check missing data
print("Missing values per column:")
missing_data = df_cleaned.isnull().sum()
print(missing_data[missing_data > 0])
print(f"\nTotal rows: {len(df_cleaned)}")
print(f"\nMissing data percentages:")
print((df_cleaned.isnull().sum() / len(df_cleaned) * 100).round(2)[df_cleaned.isnull().sum() > 0])
```

```python
# Compute sector medians and fill missing room values
rooms_sector_medians = df_cleaned.groupby('sector_oficial')['rooms'].median()
df_cleaned['rooms'] = df_cleaned['rooms'].fillna(df_cleaned['sector_oficial'].map(rooms_sector_medians))

# Set missing values of the "elevator" column to false
df_cleaned['elevator'] = df_cleaned['elevator'].fillna(False)

# Set missing values of the "floors" column to 0
df_cleaned['floors'] = df_cleaned['floors'].fillna(0)

# Compute sector medians and fill missing cadastre values
cadastre_sector_medians = df_cleaned.groupby('sector_oficial')['metres_cadastre'].median()
df_cleaned['metres_cadastre'] = df_cleaned['metres_cadastre'].fillna(df_cleaned['sector_oficial'].map(cadastre_sector_medians))

# Compute sector medians and fill missing co2 emissions values
co2_sector_medians = df_cleaned.groupby('sector_oficial')['emissions_de_co2'].median()
df_cleaned['emissions_de_co2'] = df_cleaned['emissions_de_co2'].fillna(df_cleaned['sector_oficial'].map(co2_sector_medians))

# Compute sector medians and fill missing energy values
energia_map = {
    'A': 7,
    'B': 6,
    'C': 5,
    'D': 4,
    'E': 3,
    'F': 2,
    'G': 1
}

energia_map_inv = {v: k for k, v in energia_map.items()}
df_cleaned['qual_energia_num'] = df_cleaned['qual_energia'].map(energia_map)

energia_sector_medians = (
    df_cleaned
    .groupby('sector_oficial')['qual_energia_num']
    .median()
    .round()
)

df_cleaned['qual_energia_num'] = df_cleaned['qual_energia_num'].fillna(
    df_cleaned['sector_oficial'].map(energia_sector_medians)
)

df_cleaned['qual_energia'] = df_cleaned['qual_energia_num'].map(energia_map_inv)
df_cleaned.drop(columns='qual_energia_num', inplace=True)
```

```python
# Remove outlier rows, using the Interquartile Range (IQR) method for outlier detection
# Logarithmic transformation is applied to skewed columns before outlier detection
# Linear transformation is applied to columns that are not skewed before outlier detection
def iqr_outliers(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return series[(series < lower) | (series > upper)]

def group_linear_iqr_outliers(df, value_col):
    outlier_idx = []
    for _, g in df.groupby('sector_oficial'):
        outliers = iqr_outliers(g[value_col])
        outlier_idx.extend(outliers.index)
    return df.loc[outlier_idx]

def group_log_iqr_outliers(df, value_col):
    outlier_idx = []
    for _, g in df.groupby('sector_oficial'):
        log_vals = np.log1p(g[value_col])
        outliers = iqr_outliers(log_vals)
        outlier_idx.extend(outliers.index)
    return df.loc[outlier_idx]

linear_iqr_cols = [
    'rooms',
    'floors',
    'emissions_de_co2'
]

log_iqr_cols = [
    'price',
    'area',
    'metres_cadastre'
]

outlier_indices = set()

# Linear-IQR columns
for col in linear_iqr_cols:
    outliers = group_linear_iqr_outliers(df_cleaned, col)
    outlier_indices.update(outliers.index)

# Log-IQR columns
for col in log_iqr_cols:
    outliers = group_log_iqr_outliers(df_cleaned, col)
    outlier_indices.update(outliers.index)

df_no_outliers = (
    df_cleaned
    .drop(index=outlier_indices)
    .reset_index(drop=True)
)

df_no_outliers 
```

```python
# Save the cleaned dataset to a new CSV file
df_no_outliers.to_csv('dataset_barris_cleaned.csv', index=False)
```
